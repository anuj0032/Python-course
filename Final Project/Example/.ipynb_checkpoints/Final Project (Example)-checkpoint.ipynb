```python
df_born.head()
```

#### Hvor fødes det flest barn?

```python

fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

sns.barplot(x='fylke',y='fødte',data=df_born, ax=ax, )
```

```python


fig, ax = plt.subplots()
fig.set_size_inches(5, 10)
ax.set_title('Antall fødte i hvert fylke (2.kvartal 2018)')

sns.barplot(x='fødte',y='fylke',data=df_born, ax=ax, )
```

# Ønsker å undersøke befolkningsvekst

```python
df.head(10)
```

```python
population_growth_norway = df[(df['statistikkvariabel']=='Folkevekst')]

population_growth_norway['fylke'] = population_growth_norway['region'].apply(lambda x: x.split()[1])
```

```python
population_growth_norway.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

sns.barplot(x='fylke',y='Folkemengd og kvartalsvise endringar, etter region, kvartal og statistikkvariabel',data=population_growth_norway, ax=ax, )
```

```python
population_total_norway = df[ (df['statistikkvariabel'] =='Folketalet ved inngangen av kvartalet')]

#create a column that is easier to work with (more pretty version of region - without number in front)
population_total_norway['fylke'] = population_total_norway['region'].apply(lambda x: x.split()[1])
```

```python
population_total_norway.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

sns.barplot(x='fylke',y='Folkemengd og kvartalsvise endringar, etter region, kvartal og statistikkvariabel',data=population_total_norway, ax=ax, )
```

```python
merged = pd.merge(population_total_norway,population_growth_norway,on='fylke' ) 
```

```python
merged.head()
```

```python
population_total_norway['vekst'] = population_growth_norway['Folkemengd og kvartalsvise endringar, etter region, kvartal og statistikkvariabel']
```

```python
merged['vekst%'] = merged['Folkemengd og kvartalsvise endringar, etter region, kvartal og statistikkvariabel_y']/merged['Folkemengd og kvartalsvise endringar, etter region, kvartal og statistikkvariabel_x']
```

```python
merged['vekst i prosent'] = merged['vekst%']*100
```

```python
merged['vekst i prosent']
```

```python
sortert = merged.sort_values('vekst i prosent', ascending=False)
```

#### I hvilke fylker er veksten størst?

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

sns.barplot(x='fylke',y='vekst i prosent',data=sortert, ax=ax, )
```

#### Hvordan ser fordeling av befolkning ut i forhold til hvor veksten kommer hen?

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

ax.set_title('Fordeling av befolkning per fylke')
sns.barplot(x='fylke',y='Folkemengd og kvartalsvise endringar, etter region, kvartal og statistikkvariabel_x',data=merged, ax=ax, )
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)

ax.set_title('Vekst i prosent årlig')
sns.barplot(x='fylke',y='vekst i prosent',data=merged, ax=ax, )
```
