Plot the function f(x) = 2x + 8 for x=[0,100]

** Remember to set axis labels and Titles

```python
# Write code here..

y = 2*x + 8

plt.plot(x,y)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Title')

plt.show()
```

#### Excerise 4

Plot how the BMI varies on a person with a height of 1.75 meter

**Formula: **
\begin{align}
BMI & = \frac{weight[kg]}{height[m]^2} 
\end{align}

```python
# Write code here..

# BMI(weight,height=constant) --> BMI(weight) = weight / 1.75**2

bmi = x/1.75**2

plt.plot(x,bmi)

plt.xlabel('Weight')
plt.ylabel('BMI(weight)')
plt.title('Linear relation of bmi and weight')

plt.show()
```
