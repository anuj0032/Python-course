#### Excerise 6 

Make a new columns that give a Body Mass index category to every person

BMI categories:

Underweight: 16-18.5

Normal: 18.5-25

Overweight 25-30

Obese  >30

```python
# Write code here..

def bmi_category(bmi):
    
    if (bmi > 16) and (bmi <=18.15):
        return 'Underweight'
    
    elif (bmi > 18.5) and (bmi <=25):
        return 'Normal weight'

    elif (bmi > 25) and (bmi <=30):
        return 'Overweight'
    
    elif (bmi > 30):
        return 'Obese'
    
    else:
        return 'Invalid'

df['bmi_category'] = df['bmi'].apply(bmi_category)
```
