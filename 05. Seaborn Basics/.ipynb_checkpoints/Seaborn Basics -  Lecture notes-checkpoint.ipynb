## Load dataset 

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

___

<h3 align="center">Distribution plots</h3> 

### distplot ( )

Read more: https://seaborn.pydata.org/generated/seaborn.distplot.html

```python
sns.distplot(tips['total_bill'])
```

### jointplot ( )

Plot the distribution of two variables using jointplot.

Read more: https://seaborn.pydata.org/generated/seaborn.jointplot.html?highlight=jointplot#seaborn.jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

### pairplot ( )

Read more: https://seaborn.pydata.org/generated/seaborn.pairplot.html?highlight=pairplot#seaborn.pairplot

```python
sns.pairplot(tips)
```

___

<h3 align="center">Categorical plots</h3> 

### barplot ( )

Read more: https://seaborn.pydata.org/generated/seaborn.barplot.html?highlight=barplot#seaborn.barplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

### countplot ( )

Read more: https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
sns.countplot(x='sex',data=tips)
```

### boxplot ( )

Read more: https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

___

<h3 align="center">Matrix plot</h3> 

### heatmap ( )

Read more: https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
# Matrix form for correlation data
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

___

<h3 align="center">Linear model plots</h3> 

### lmplot ( )

Read more: https://seaborn.pydata.org/generated/seaborn.lmplot.html

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```
